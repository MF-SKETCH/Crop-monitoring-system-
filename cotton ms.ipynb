```python
import random
import matplotlib.pyplot as plt

# -------------------------------
# Part 1 – Setup
# -------------------------------
n_days = 5
temperature = []
moisture = []
rainfall = []

# -------------------------------
# Part 2 – Generate Data
# -------------------------------
for day in range(1, n_days + 1):
    temp = random.randint(18, 40)
    moist = random.randint(20, 90)
    rain = random.randint(0, 100)

    temperature.append(temp)
    moisture.append(moist)
    rainfall.append(rain)

    print(f"Day {day}: Temp={temp}°C | Moisture={moist}% | Rain={rain}mm")

# -------------------------------
# Part 3 – Plot Graph
# -------------------------------
days = list(range(1, n_days + 1))  # X-axis

# Final check – lengths must match
print("Length Check:")
print(len(days), len(temperature), len(moisture), len(rainfall))

plt.plot(days, temperature, label="Temperature (°C)")
plt.plot(days, moisture, label="Soil Moisture (%)")
plt.plot(days, rainfall, label="Rainfall (mm)")

plt.xlabel("Days")
plt.ylabel("Values")
plt.title("Cotton Crop Monitoring")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Length and type check:")
print("Days:", len(days), type(days))
print("Temperature:", len(temperature), type(temperature))
print("Moisture:", len(moisture), type(moisture))
print("Rainfall:", len(rainfall), type(rainfall))
```
