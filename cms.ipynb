```python
import random
import matplotlib.pyplot as plt

# --- Function 1: Generate random data ---
def generate_data(days=7):
    """Random data generation for given days"""
    temperature = [random.randint(15, 45) for _ in range(days)]
    soil_moisture = [random.randint(10, 100) for _ in range(days)]
    rainfall = [random.randint(0, 50) for _ in range(days)]
    return temperature, soil_moisture, rainfall

# --- Function 2: Analyze data ---
def analyze_data(temperature, soil_moisture, rainfall):
    """Analyze and give suggestions"""
    print("\n--- Crop Monitoring Report ---")
    for day in range(len(temperature)):
        print(f"\nDay {day+1}:")
        print(f" Temperature: {temperature[day]} °C")
        print(f" Soil Moisture: {soil_moisture[day]} %")
        print(f" Rainfall: {rainfall[day]} mm")

        # Soil moisture condition
        if soil_moisture[day] < 30:
            print(" Suggestion: Irrigation required.")
        else:
            print(" Soil moisture is good.")

        # Temperature condition
        if temperature[day] > 35:
            print(" High temp → Risk of heat stress.")
        elif temperature[day] < 20:
            print(" Low temp → Crop growth may slow down.")
        else:
            print(" Temperature is optimal.")

        # Rainfall condition
        if rainfall[day] < 10:
            print(" Rainfall too low → Irrigation may be needed.")
        else:
            print(" Rainfall is adequate.")

# --- Function 3: Plot graph ---
def plot_data(temperature, soil_moisture, rainfall):
    """Plot graphs of crop data"""
    days = list(range(1, len(temperature)+1))

    plt.figure(figsize=(10,6))
    plt.plot(days, temperature, marker='o', label="Temperature (°C)")
    plt.plot(days, soil_moisture, marker='s', label="Soil Moisture (%)")
    plt.plot(days, rainfall, marker='^', label="Rainfall (mm)")

    plt.title("Crop Monitoring System")
    plt.xlabel("Days")
    plt.ylabel("Values")
    plt.legend()
    plt.grid(True)
    plt.show()

# --- Main Program ---
def main():
    temperature, soil_moisture, rainfall = generate_data()
    analyze_data(temperature, soil_moisture, rainfall)
    plot_data(temperature, soil_moisture, rainfall)

# Run the project
if __name__ == "__main__":
    main()
```
